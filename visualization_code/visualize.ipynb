```python
import matplotlib.pyplot as plt

voltage_data = []

file_name = './recordings/test2.txt'

# Read the voltage values from the text file, skipping empty lines and non-numeric values
with open(file_name, 'r') as file:
    for line in file:
        line = line.strip()
        if line:
            try:
                voltage = float(line)
                voltage_data.append(voltage-2.5)
            except ValueError:
                print(f"Skipping invalid data: {line}")

# Create a time (x-axis) array based on the number of data points
time = range(len(voltage_data))

# Create a line plot
plt.figure(figsize=(10, 2))
plt.plot(time, voltage_data)
plt.xlabel('Sample Index')
plt.ylabel('Voltage (V)')
plt.title('Voltage vs. Sample Index')
plt.xlim(2000, 2300)
plt.grid(True)

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

voltage_data = []

file_name = './recordings/test2.txt'
sampling_rate = 800  # Replace with your actual sampling rate in Hz

# Read the voltage values from the text file, skipping empty lines and non-numeric values
with open(file_name, 'r') as file:
    for line in file:
        line = line.strip()
        if line:
            try:
                ADC_val = float(line)
                # Convert to voltage (5V reference)
                voltage_val = ADC_val * (5 / 1023)
                # subtract the DC shift
                voltage_data.append(voltage_val - 2.5)
            except ValueError:
                print(f"Skipping invalid data: {line}")

# Convert voltage_data to a NumPy array
voltage_data = np.array(voltage_data)

# Calculate the time vector based on the sampling rate
time = np.arange(len(voltage_data)) / sampling_rate

# Mean normalize the voltage data
voltage_data = voltage_data - np.mean(voltage_data)

# Create a line plot
plt.figure(figsize=(10, 2))
plt.plot(time, voltage_data)
plt.xlabel('Time (s)')
plt.ylabel('Voltage (V)')
plt.title('Voltage vs. Time')
plt.xlim(2, 2.3)  # Adjust the time range as needed
plt.grid(True)

# Display the plot
plt.show()
```
